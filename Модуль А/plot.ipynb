# 1. Создаем граф плотности распределения данных

```python
# Импорттруем библиотеку для работы с данными
import pandas as pd
# ИМпортиурем библиотеку для работы с графами
import matplotlib.pyplot as plt

# Читаем данные из CSV в DataFrame
data = pd.read_csv(r'C:\Users\pupki\Desktop\BiAnlysisForTime1\Модуль А\data\cleardata\csvcleardata.csv', encoding='windows-1251')

# Создаем график плотности распределения для столбца salary
data.plot(kind='density')

# Добавляем заголовок и метки осей
plt.title('Плотность распределения данных')
plt.xlabel('Значение')
plt.ylabel('Плотность')

# Отображаем график
plt.show()
```
