# 1. Реализация корреляционного анализа

```python
# Импортиурем библиотеку для работы с данными
import pandas as pd
# Импортиурем модуль из бибилиотеки для работы со статистическими методами
from scipy.stats import spearmanr
```

```python
# Загружаем данные из выгруженного csv в дата фрейм
df = pd.read_csv('csvcleardata.csv', encoding='utf-8')

# Выводим пример данных
df
```

```python
# Выводим корреляционный анализ переменных: заработная плата и опыт работы
print(f"Корреляционный анализ переменных: заработная плата и опыт работы - {spearmanr(df['salary'], df['work_experience'])}", end='\n\n')
# Выводим корреляционный анализ переменных: заработная плата и регион
print(f"Корреляционный анализ переменных: заработная плата и регион - {spearmanr(df['salary'], df['region_name'])}")
```

### Объяснение полученного вывода
statistic - указывает на линейную зависимость между переменными в диапозоне от -1 до 1, чем ближе к одному, тем больше зависимость между переменными
